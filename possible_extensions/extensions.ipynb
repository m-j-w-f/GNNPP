## Bernstein Quantile Network

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

plt.style.use('default')
plt.rcParams.update({'font.size': 12})
%matplotlib inline
```

```python
# Define the degree of the Bernstein polynomial
d = 8

# Generate values of t in the range [0, 1]
t_values = np.linspace(0, 1, 1000)

# Create the plot
plt.figure(figsize=(10, 6))
ax = plt.axes()
ax.set_prop_cycle('color',[plt.cm.jet(i) for i in np.linspace(0, 1, d+1)])

all_bernstein_values = []
# Calculate the Bernstein polynomial values for each t
for l in range(d + 1):
    # Initialize an empty list to store the polynomial values
    bernstein_values = []
    for t in t_values:
        bernstein = comb(d, l) * t**l * (1 - t)**(d - l)
        bernstein_values.append(bernstein)
    all_bernstein_values.append(bernstein_values)
    # Create the plot
    name = '$' + 'B_' + '{' + str(l) + ',' + str(d)+ '}' + '$'
    ax.plot(t_values, bernstein_values, label=name)


plt.xlabel('probability')
plt.title(f'Bernstein Polynomial Basis Functions of Degree {d}')
plt.grid(True)
plt.legend(ncols=3)
plt.savefig(f'Bernstein_Polynomial_Basis_Functions_of_Degree_{d}.pdf', format='pdf', dpi=600, bbox_inches='tight')
```

```python
# Random Linear Combination
bernstein_np = np.array(all_bernstein_values)

plt.figure(figsize=(10, 6))
ax = plt.axes()
ax.set_prop_cycle('color',[plt.cm.jet(i) for i in np.linspace(0, 1, d+1)])

for _ in range(10):
    random_numbers = np.random.normal(2,40,d+1)
    sorted_numbers = np.sort(random_numbers)
    linear_combination = sorted_numbers.dot(bernstein_np)

    ax.plot(t_values, linear_combination)
```
